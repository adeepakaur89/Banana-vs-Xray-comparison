```python
import matplotlib.pyplot as plt
import numpy as np

# Banana dose per banana
banana_dose = 0.1  # microSieverts per banana

# Banana counts to show
banana_counts = np.array([1, 50, 100, 500, 1000])

# Dose values
banana_doses = banana_counts * banana_dose

# Chest X-ray dose
xray_dose = 100  # microSieverts

plt.figure(figsize=(10,6))

# Bar graph in yellow
plt.bar([str(i) for i in banana_counts], banana_doses, label="Banana Dose", color="yellow")

# Horizontal line for chest X-ray
plt.axhline(y=xray_dose, linestyle='--', linewidth=2, label="Chest X-Ray Dose (100 µSv)")

# Labels
plt.title("Radiation Dose: Bananas vs Chest X-Ray")
plt.xlabel("Number of Bananas")
plt.ylabel("Dose (microSieverts, µSv)")

# Legend
plt.legend()

plt.tight_layout()
plt.show()
```
